#**Theoretical Questions:**

#1. What is hypothesis testing in statistics?
- Statistical hypothesis testing is a method used to determine if there's enough evidence in a sample to draw conclusions about a population, involving formulating hypotheses (null and alternative), collecting data, and assessing the evidence to decide whether to reject or fail to reject the null hypothesis.

#2. What is the Null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis is a fundamental concept in statistical hypothesis testing. It represents the default or baseline assumption, suggesting that there is no effect, no difference, or no relationship between variables in a population. For example, it might state, "This treatment has no impact."

 In contrast, the alternative hypothesis is what researchers aim to support, proposing that there is an effect, a difference, or a relationship between variables. For instance, it might argue, "This treatment significantly improves outcomes."   

 To summarize:

 - Null hypothesis (H‚ÇÄ): The assumption of no change or effect.

 - Alternative hypothesis (H‚ÇÅ): The assumption of a real change or effect.

#3. What is the significance level in hypothesis testing, and why is it important?
- The significance level, often denoted as Œ± (alpha), is a threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the probability of making a Type I error‚Äîrejecting the null hypothesis when it is actually true. Commonly used significance levels are 0.05, 0.01, or 0.10.

 **Why is it important?**     
 Decision-making: It provides a criterion for determining whether the observed data is sufficiently unlikely under the null hypothesis.

 - **Balancing errors:** Lower significance levels reduce the risk of Type I errors but increase the risk of Type II errors (failing to reject a false null hypothesis).

 - **Scientific rigor:** Setting a significance level in advance prevents bias in interpreting results.

#4. What does a P-value represent in hypothesis testing?
 - A P-value is a measure used in hypothesis testing to determine the strength of evidence against the null hypothesis. It represents the probability of observing the test statistic (or something more extreme) under the assumption that the null hypothesis is true`1` `2`.

#5. How do you interpret the P-value in hypothesis testing?
 - Interpreting the P-value in hypothesis testing involves understanding its role as a measure of evidence against the null hypothesis. Here's how to interpret it:

  - **Small P-value (e.g., ‚â§ 0.05):** Indicates strong evidence against the null hypothesis, suggesting it may be rejected. For instance, if the P-value is 0.03, there's only a 3% probability of observing the data (or something more extreme) if the null hypothesis is true.

  - **Large P-value (e.g., > 0.05):** Suggests weak evidence against the null hypothesis, meaning it may not be rejected. This implies the data is consistent with the null hypothesis.

#6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type 1 error(False Positive):      
   - It occurs when the null hypothesis is rejected even though it is true.
   - The probability of making a Type 1 error is equal to the significance level (Œ±).

 - Type 2 error(False Negative):      
   - It happens when the null hypothesis is not rejected even though it is false.
   - The probability of making a Type 2 error is donated by Œ≤.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 - Difference between a one-tailed test and a two-tailed test in hypothesis testing:             

  **One-Tailed Test**:
  - Tests whether the effect or relationship exits in a specific direction(e.g., greater than or less than a certain value).
  - It located in only one tail of the distribution(either the left tail or the right tail).

  **Two-Tailed Test**:             
  - Tests whether the effect or relationship exits in either direction(e.g., different from a certain value, whether it's greater or smaller).
  - It splits between two tails of the distribution.

#10. What is the T-distribution, and when should it be used instead of the normal distribution?
 - The T-distribution (or Student's T-distribution) is a probability distribution used in statistics, particularly in hypothesis testing and confidence interval estimation. It is similar to the normal distribution but has heavier tails, meaning it accounts for more variability in the data.

**When should it be used instead of the normal distribution?**

1. **Small Sample Sizes:** When the sample size is small (typically n < 30), the T-distribution is more appropriate because it adjusts for the increased uncertainty in estimating the population standard deviation.

2. **Unknown Population Standard Deviation:** If the population standard deviation is unknown and must be estimated from the sample data, the T-distribution is used.

3. **Robustness:** It is more robust for small datasets or when the data slightly deviates from normality.

#12. What is the T-test, and how is it used in hypothesis testing?
 - The T-test is a statistical method used in hypothesis testing to determine whether there is a significant difference between the means of two groups or between a sample mean and a known value. It is particularly useful when the sample size is small and the population standard deviation is unknown.

**How is it used?**

1. **Formulate Hypotheses:** Define the null hypothesis (no difference) and the alternative hypothesis (a significant difference).

2. **Calculate the T-statistic:** Use the sample data to compute the T-statistic, which measures the difference relative to the variability in the data.

3. **Compare to Critical Value or P-value:** It determine whether the T-statistic falls within the critical region or whether the P-value is below the significance level (Œ±).

4. **Make a Decision:** It reject or fail to reject the null hypothesis based on the results.

#13. What is the relationship between Z-test and T-test in hypothesis testing?
 - The Z-test and T-test are closely related statistical methods used in hypothesis testing, but they differ in their assumptions and applications. Here's how they are connected:

1. **Purpose:** Both tests aim to determine whether there is a significant difference between sample data and population parameters or between two groups.

2. **Distribution:** The Z-test relies on the standard normal distribution, while the T-test uses the T-distribution, which accounts for more variability in small samples.

3. **Sample Size:** The Z-test is typically used for large samples (n ‚â• 30), whereas the T-test is preferred for small samples (n < 30).

4. **Population Variance:** The Z-test requires the population variance to be known, while the T-test is used when the population variance is unknown and estimated from the sample.






#14. What is a confidence interval, and how is it used to interpret statistical results?
 - A **confidence interval** is a range of values derived from sample data that is likely to contain the true population parameter (e.g., mean or proportion). It provides a measure of uncertainty in an estimate and is expressed with a confidence level, such as 95% or 99%.-

**How It's Used:**
1. **Interpretation:** A 95% confidence interval means that if we were to take 100 different samples and calculate intervals for each, approximately 95 of those intervals would contain the true population parameter.

2. **Decision-Making:** Confidence intervals help researchers assess the reliability of their estimates and make informed decisions.

3. **Precision:** Narrow intervals indicate more precise estimates, while wider intervals suggest greater uncertainty.

#15. What is the margin of error, and how does it affect the confidence interval?
 - The margin of error quantifies the uncertainty in a statistical estimate, representing the maximum expected difference between the sample statistic and the true population parameter. It is influenced by factors like sample size, variability, and confidence level.

**How It Affects Confidence Interval:**   

1. **Width of Interval:** The margin of error determines the range of the confidence interval. A larger margin of error results in a wider interval, indicating greater uncertainty, while a smaller margin of error leads to a narrower interval, suggesting more precision.
2. **Confidence Level:** Higher confidence levels (e.g., 99% vs. 95%) increase the margin of error, as they require a broader interval to ensure the true parameter is captured.

3. **Sample Size:** Larger sample sizes reduce the margin of error, leading to more precise confidence intervals.

#19. What is the F-distribution, and when is it used in hypothesis testing?

#20. What is an ANOVA test, and what are its assumptions?
 - The ANOVA test (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if there are significant differences among them. It analyzes the variability within and between groups to assess whether observed differences are due to chance or reflect true effects.

**Assumptions of ANOVA:**

1. **Normality:** The data in each group should be drawn from a normally distributed population.

2. **Homogeneity of Variances:** The variances of the populations being compared should be equal.

3. **Independence:** Observations within and between groups must be independent of each other.

#21. What are the different types of ANOVA tests?
 - ANOVA tests come in various types, each suited to specific scenarios:

1. **One-Way ANOVA:** Compares the means of three or more groups based on a single independent variable. For example, testing the effect of different fertilizers on plant growth.

2. **Two-Way ANOVA:** Examines the impact of two independent variables simultaneously, including their interaction effects. For instance, analyzing how teaching methods and class timings affect student performance.

3. **Repeated Measures ANOVA:** Used when the same subjects are tested under different conditions or time points, controlling for individual variability. For example, measuring the effect of a drug on patients over multiple time intervals.

#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import scipy.stats as stats

sample_mean = float(input("Enter the sample mean: "))
population_mean = float(input("Enter the population mean: "))
population_std_dev = float(input("Enter the population standard deviation: "))
sample_size = int(input("Enter the sample size: "))

z_score = (sample_mean - population_mean) / (population_std_dev / (sample_size ** 0.5))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?


```python
import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample, population_mean, population_std_dev):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

np.random.seed(42)
sample = np.random.normal(loc=75, scale=10, size=30)
population_mean = 75
population_std_dev = 10

z_score, p_value = one_sample_z_test(sample, population_mean, population_std_dev)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, population_mean, population_std_dev, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    critical_value = stats.norm.ppf(1 - alpha / 2)

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value, critical_value

np.random.seed(42)
sample_size = 30
population_mean = 75
population_std_dev = 10
random_sample = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

z_score, p_value, critical_value = two_tailed_z_test(random_sample, population_mean, population_std_dev)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y, label="Standard Normal Distribution")
plt.axvline(-critical_value, color='red', linestyle='--', label=f"Critical Value (-{critical_value:.2f})")
plt.axvline(critical_value, color='red', linestyle='--', label=f"Critical Value ({critical_value:.2f})")
plt.axvline(z_score, color='blue', linestyle='-', label=f"Z-score ({z_score:.2f})")
plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label="Rejection Region")
plt.title("Decision Region for Two-Tailed Z-Test")
plt.legend()
plt.xlabel("Z-values")
plt.ylabel("Probability Density")
plt.show()

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference found.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_and_type2_errors(mu0, mu1, sigma, sample_size, alpha):

    standard_error = sigma / np.sqrt(sample_size)

    z_critical = stats.norm.ppf(1 - alpha)
    critical_value = mu0 + z_critical * standard_error

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y_null = stats.norm.pdf(x, loc=mu0, scale=standard_error)
    y_alternate = stats.norm.pdf(x, loc=mu1, scale=standard_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color='blue')
    plt.plot(x, y_alternate, label="Alternate Hypothesis (H1)", color='green')

    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.4, label="Type 1 Error (Œ±)")

    beta_region = x <= critical_value
    plt.fill_between(x, y_alternate, where=beta_region, color='orange', alpha=0.4, label="Type 2 Error (Œ≤)")

    plt.axvline(x=critical_value, color='black', linestyle='--', label=f"Critical Value = {critical_value:.2f}")

    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

    beta = stats.norm.cdf(critical_value, loc=mu1, scale=standard_error)
    print(f"Type 2 Error (Œ≤): {beta:.4f}")
    print(f"Power of the Test (1 - Œ≤): {1 - beta:.4f}")

mu0 = 100
mu1 = 105
sigma = 15
sample_size = 30
alpha = 0.05

visualize_type1_and_type2_errors(mu0, mu1, sigma, sample_size, alpha)
```

#6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The means of the two samples are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample means.")

np.random.seed(42)
sample1 = np.random.normal(loc=75, scale=10, size=30)
sample2 = np.random.normal(loc=80, scale=12, size=30)

independent_t_test(sample1, sample2)
```

#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)
    standard_error = sample_std / np.sqrt(sample_size)

    degrees_of_freedom = sample_size - 1
    t_critical = t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

    margin_of_error = t_critical * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=30)
confidence_level = 0.95

lower_bound, upper_bound = confidence_interval(sample_data, confidence_level)
print(f"{confidence_level*100:.0f}% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)

    standard_error = sample_std / np.sqrt(sample_size)

    degrees_of_freedom = sample_size - 1
    t_critical = t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

    margin_of_error = t_critical * standard_error

    return margin_of_error

np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=30)
confidence_level = 0.95

margin_of_error = calculate_margin_of_error(sample_data, confidence_level)
print(f"Margin of Error for {confidence_level*100:.0f}% confidence level: {margin_of_error:.2f}")
```

#16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    p_value = 1 - f.cdf(f_stat, df1, df2)
    p_value = p_value * 2

    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

f_test(sample1, sample2)
```

#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt

def check_anova_assumptions(data, groups_column, response_column):
    grouped_data = data.groupby(groups_column)[response_column]

    print("Normality Test Results:")
    for group_name, group_data in grouped_data:
        stat, p_value = shapiro(group_data)
        print(f"Group '{group_name}': W-statistic={stat:.4f}, P-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"  -> Group '{group_name}' does not follow a normal distribution (P < 0.05).")
        else:
            print(f"  -> Group '{group_name}' follows a normal distribution (P >= 0.05).")

    group_values = [group_data.values for _, group_data in grouped_data]
    stat, p_value = levene(*group_values)
    print("\nEqual Variance Test Results (Levene's Test):")
    print(f"Levene's Statistic={stat:.4f}, P-value={p_value:.4f}")
    if p_value < 0.05:
        print("  -> Variances are not equal among groups (P < 0.05).")
    else:
        print("  -> Variances are equal among groups (P >= 0.05).")

    print("\nIndependence Assumption:")
    print("Independence must be ensured during data collection. This assumption cannot be tested using statistical methods.")

    plt.figure(figsize=(8, 6))
    data.boxplot(column=response_column, by=groups_column)
    plt.title("Boxplot of Response by Groups")
    plt.suptitle("")
    plt.xlabel(groups_column)
    plt.ylabel(response_column)
    plt.grid(True)
    plt.show()

np.random.seed(42)
data = pd.DataFrame({
    'Group': np.repeat(['A', 'B', 'C'], 20),
    'Response': np.concatenate([
        np.random.normal(50, 5, 20),
        np.random.normal(55, 5, 20),
        np.random.normal(60, 5, 20)
    ])
})

check_anova_assumptions(data, groups_column='Group', response_column='Response')
```

#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2

def test_population_variance(sample, hypothesized_variance, alpha=0.05):
    sample_size = len(sample)
    sample_variance = np.var(sample, ddof=1)

    chi_square_stat = (sample_size - 1) * sample_variance / hypothesized_variance

    df = sample_size - 1

    critical_value_low = chi2.ppf(alpha / 2, df)
    critical_value_high = chi2.ppf(1 - alpha / 2, df)

    p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Chi-square Statistic: {chi_square_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"Critical Values: ({critical_value_low:.4f}, {critical_value_high:.4f})")
    print(f"P-value: {p_value:.4f}")

    if chi_square_stat < critical_value_low or chi_square_stat > critical_value_high:
        print("Reject the null hypothesis: The variance significantly differs from the hypothesized variance.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variance.")

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)
hypothesized_variance = 100

test_population_variance(sample, hypothesized_variance)
```

#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from scipy.stats import norm

def z_test_proportions(successes1, size1, successes2, size2, alpha=0.05):
    p1 = successes1 / size1
    p2 = successes2 / size2

    p_pool = (successes1 + successes2) / (size1 + size2)

    standard_error = np.sqrt(p_pool * (1 - p_pool) * (1 / size1 + 1 / size2))

    z_score = (p1 - p2) / standard_error

    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Proportion 1: {p1:.4f}")
    print(f"Proportion 2: {p2:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the proportions.")

successes1 = 45
size1 = 100
successes2 = 30
size2 = 90

z_test_proportions(successes1, size1, successes2, size2)
```

#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_stat = var1 / var2 if var1 >= var2 else var2 / var1
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    critical_value_low = f.ppf(alpha / 2, df1, df2)
    critical_value_high = f.ppf(1 - alpha / 2, df1, df2)

    p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

    return f_stat, p_value, critical_value_low, critical_value_high, df1, df2

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=60, scale=10, size=30)

alpha = 0.05
f_stat, p_value, critical_value_low, critical_value_high, df1, df2 = f_test(sample1, sample2, alpha)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: df1={df1}, df2={df2}")
print(f"Critical Values: ({critical_value_low:.4f}, {critical_value_high:.4f})")

if f_stat < critical_value_low or f_stat > critical_value_high:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

x = np.linspace(0, 5, 500)
y = f.pdf(x, df1, df2)
plt.plot(x, y, label=f"F-distribution (df1={df1}, df2={df2})")
plt.axvline(f_stat, color='blue', linestyle='--', label=f"F-statistic ({f_stat:.4f})")
plt.axvline(critical_value_low, color='red', linestyle='--', label=f"Critical Value Low ({critical_value_low:.4f})")
plt.axvline(critical_value_high, color='red', linestyle='--', label=f"Critical Value High ({critical_value_high:.4f})")
plt.fill_between(x, y, where=(x < critical_value_low) | (x > critical_value_high), color='red', alpha=0.2, label="Rejection Region")
plt.title("F-Test: Comparing Variances")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```
